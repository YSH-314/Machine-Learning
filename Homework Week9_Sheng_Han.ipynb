# 1. Implement k-NN method (50 Points)

```python
import numpy as np
import scipy.io
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats as st
```

```python
def KNN(train_x, train_y, test_x, k):
    predict = []
    for row in range(test_x.shape[0]):
        distance=[]
        for i in range(train_x.shape[0]):
            distance.append(np.linalg.norm(train_x[i,:]-test_x[row,:],2))
        arr = np.array(distance)
        index = arr.argsort()[:k]
        #print(train_y[index].reshape(-1).tolist())
        predict.append(st.mode(train_y[index].reshape(-1).tolist())[0][0])
    return predict   
```

# 2. Load train and test mat files, perform k-NN and report acuracy on the test dataset (30 Points)

```python
train = scipy.io.loadmat('train.mat')
test = scipy.io.loadmat('test.mat')
```

```python
train_x = train['features']
train_y = train['labels']
test_x = test['features']
test_y = test['labels']
```

```python
def acc(train_y, pred):
    correct=0
    for i in range(len(pred)):
        if pred[i]==test_y[i]:
            correct+=1
    acc= round(correct/len(pred)*100,2)
    return acc
```

```python
acc_list=[]
for k in range(1,16,2):
    pred = KNN(train_x, train_y, test_x, k)
    acc_list.append(acc(train_y, pred))
```

# 3. Reduce the dimensionality of features using PCA to low dimensions (e.g., 10, 3 ) and report the accuracy of the test datasets (20 Points)

```python
def PCA(A, num_dimension):
    
    # Transfer A with shape (# of featurs, # raws of data)
    temp = np.zeros(shape=(A.shape[1],A.shape[0]))
    for i in range(len(A)):
        temp[:,i] = A[i] 
    A = temp    
    
    # N: # of features, M: # data
    N , M = A.shape 
    # Step 1: compute sample mean
    mean_x = np.mean(A, axis=0)
    
    # Step 2: subtract sample mean (i.e., center data at zero)
    center_A = A - mean_x
    
    # Step 3: compute the sample covariance matrix Σx
    cov_A = (center_A @ center_A.T)/M
    
    # Step 4: compute the SVD of Σx
    u, s, vh = np.linalg.svd(cov_A, full_matrices=True)
    
    # Step 5: dimensionality reduction step
    reduced_u = u[:,:num_dimension]
    reduced_A = reduced_u.T @ center_A
    
    # Step 6: transfer features to columns, shape = (# data, # of features)
    res = np.zeros(shape = (reduced_A.shape[1], reduced_A.shape[0]))
    for col in range(reduced_A.shape[1]):
        res[col,:] = reduced_A[:, col]
                   
    
    return res
```

```python
num_dimension = [3, 5, 10]
acc_list = []
for num in num_dimension:
    reduced_train = PCA(train_x, num)
    print('shape of reduced_train', reduced_train.shape)
    reduced_test = PCA(test_x, num)
    print('shape of reduced_test', reduced_test.shape)
    pred = KNN(reduced_train, train_y, reduced_test, 11)
    acc_list.append(acc(train_y, pred))
        
```

```python
plt.plot(num_dimension, acc_list)
plt.xlabel('# dimensions')
plt.ylabel('accuracy(%)')
```
